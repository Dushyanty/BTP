```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/Dushyanty/BTP/main/BTP1.csv')
```

```python
dummies = pd.get_dummies(df['Composition'])
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
merged.columns 
```

```python
X=merged[['Temperature']]
```

```python
X.shape
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
```

```python
y= merged[['Hardness(HV)']]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
import xgboost as xgb
```

```python
gb = xgb.XGBRegressor()
```

```python
gb.fit(X_train,y_train)
```

```python
y_pred = gb.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error,accuracy_score
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_absolute_percentage_error(y_test,y_pred)
```

```python
gb.score(X_test,y_test)
```
